```python
import numpy as np
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
inverted_56_target = np.load('eval_inverted_56x56.npz')['target']
inverted_56_output = np.load('eval_inverted_56x56.npz')['output']
inverted_112_target = np.load('eval_inverted_112x112.npz')['target']
inverted_112_output = np.load('eval_inverted_112x112.npz')['output']
inverted_224_target = np.load('eval_inverted_224x224.npz')['target']
inverted_224_output = np.load('eval_inverted_224x224.npz')['output']

dynamic_56_target = np.load('eval_dynamic_56x56.npz')['target']
dynamic_56_output = np.load('eval_dynamic_56x56.npz')['output']
dynamic_112_target = np.load('eval_dynamic_112x112.npz')['target']
dynamic_112_output = np.load('eval_dynamic_112x112.npz')['output']
dynamic_224_target = np.load('eval_dynamic_224x224.npz')['target']
dynamic_224_output = np.load('eval_dynamic_224x224.npz')['output']

baseline_56_target = np.load('eval_baseline_56x56.npz')['target']
baseline_56_output = np.load('eval_baseline_56x56.npz')['output']
baseline_112_target = np.load('eval_baseline_112x112.npz')['target']
baseline_112_output = np.load('eval_baseline_112x112.npz')['output']
baseline_224_target = np.load('eval_baseline_224x224.npz')['target']
baseline_224_output = np.load('eval_baseline_224x224.npz')['output']

num_images = len(inverted_56_target)
```

# Inverted Dynamic Resolution

```python
class_names = ['tench', 'english springer', 'cassette player', 'chainsaw', 'church', 'french horn', 'garbage truck', 'gas pump', 'golf ball', 'parachute']

# plot inverted confusion matrices
cm_inverted_56 = confusion_matrix(inverted_56_target, inverted_56_output)
cm_inverted_112 = confusion_matrix(inverted_112_target, inverted_112_output)
cm_inverted_224 = confusion_matrix(inverted_224_target, inverted_224_output)

matrices = {'56':cm_inverted_56, '112':cm_inverted_112, '224':cm_inverted_224}

fig, axes = plt.subplots(1, 3, figsize=(24, 8))

for i, (res, cm) in enumerate(matrices.items()):
  ax = axes[i]

  sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names, ax=ax, cbar=False)

  ax.set_title(f"Inverted Dynamic Resolution Accuracy on {res}x{res} Test Set")
  ax.set_xlabel("Predicted Label")
  ax.set_ylabel("Target Label")

plt.tight_layout()
plt.show()
```

```python
print("INVERTED DYNAMIC RESOLUTION ACCURACY")
print("Overall (56x56):", (inverted_56_target==inverted_56_output).sum() / num_images)
print("Overall (112x112):", (inverted_112_target==inverted_112_output).sum() / num_images)
print("Overall (224x224):", (inverted_224_target==inverted_224_output).sum() / num_images)
```

```python
print("INVERTED DYNAMIC RESOLUTION ACCURACY")
for i in range(len(class_names)):
  class_mask = (inverted_56_target == i)
  print(f"{class_names[i].capitalize()} (56x56):", (inverted_56_target[class_mask]==inverted_56_output[class_mask]).sum() / class_mask.sum())

print("")
for i in range(len(class_names)):
  class_mask = (inverted_112_target == i)
  print(f"{class_names[i].capitalize()} (112x112):", (inverted_112_target[class_mask]==inverted_112_output[class_mask]).sum() / class_mask.sum())

print("")
for i in range(len(class_names)):
  class_mask = (inverted_224_target == i)
  print(f"{class_names[i].capitalize()} (224x224):", (inverted_224_target[class_mask]==inverted_224_output[class_mask]).sum() / class_mask.sum())
```
